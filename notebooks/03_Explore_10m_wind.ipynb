```python
import zipfile
import os

zip_file = "u wind.zip"
extract_folder = "wind-u"

with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print("Files extracted:")
print(os.listdir(extract_folder))
```

```python
ds_u = xr.open_dataset("wind-u/UGRD-10m_2020050222_ncum_imdaa_reanl_2df_22.nc")
```

```python
ds_u
```

```python
ds_u['10u']
```

```python
u = ds_u["10u"]
v = ds_v["10v"]
```

```python
print(ds_u.time.values)
```

```python
print(ds_v.time.values)
```

```python
import numpy as np

wind_speed = np.sqrt(u**2 + v**2)
```

```python
print(wind_speed)
```

```python
print(wind_speed.values)
```

Since we have two time frames, we have to select one time before plotting

```python
wind_speed_2d = wind_speed.isel(time=0, height=0)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

wind_speed_2d.plot(cmap="viridis")

plt.title("10-meter Wind Speed")
plt.show()
```
